# Sea Level Predictor - freecodecamp.org

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

df = pd.read_csv('epa-sea-level.csv')
df.head()
```

## Draw a scatter plot that shows the relation between year and the adjusted sea level

```python
fig, ax = plt.subplots()
df.plot(kind='scatter', ax=ax, x='Year', y='CSIRO Adjusted Sea Level')
```

## Find the line of best fit and draw it on the plot until 2050

```python
import numpy as np
from scipy.stats import linregress

m1, b1, *_ = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])

years1 = np.arange(df['Year'].min(), 2051)

sea_levels1 = m1 * years1 + b1
ax.plot(years1, sea_levels1, color='red', label='Line of Best Fit')
fig
```

## Find the line of best fit from the year 2000 and draw it on the plot until 2050

```python
m2, b2, *_ = linregress(
    df[df['Year'] >= 2000]['Year'],
    df[df['Year'] >= 2000]['CSIRO Adjusted Sea Level']
)

years2 = np.arange(2000, 2051)

sea_levels2 = m2 * years2 + b2
ax.plot(years2, sea_levels2, color='green', label='Line of Best Fit')
fig
```

Let's adjust the title and the labels

```python
ax.set_title('Rise in Sea Level')
ax.set_ylabel('Sea Level (inches)')
fig
```
